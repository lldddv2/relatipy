```python
import sympy as sp
import numpy as np
import einsteinpy.symbolic as ep
```

```python
r, theta, phi, x0, M = sp.symbols('r theta phi x_0 M')
r_s = sp.symbols('r_s')

A = 1 - r_s / r
B = -1 / A
C = -r**2
D = -r**2 * sp.sin(theta)**2

metric = np.diag(
    [A, B, C, D]
)

g = ep.MetricTensor(metric, [x0, r, theta, phi])
g.tensor()
```

```python
print(ep.ChristoffelSymbols.from_metric(g).tensor())
```
