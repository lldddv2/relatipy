```python
kerr_metric = rn.metrics.Kerr(mass=2, a=0.5)
```

```python
import time
```

```python
xs = [
        np.random.uniform(0, 10),  # t
        np.random.uniform(2.1, 10),  # r
        np.random.uniform(0, np.pi),  # theta
        np.random.uniform(0, 2 * np.pi),  # phi
    ]

# generamos 100000 conjuntos de xs
xs_list = []
for _ in range(10000):
    xs = [
        np.random.uniform(0, 10),        # t
        np.random.uniform(2.1, 10),     # r
        np.random.uniform(0, np.pi),    # theta
        np.random.uniform(0, 2 * np.pi) # phi
    ]
    xs_list.append(xs)
```

```python
# evaluar el tiempo 
t0 = time.perf_counter()
for i in range(10000):
    # random points
    xs = xs_list[i]

    christoffel = kerr_metric.get_christoffel_symbols(xs)
t1 = time.perf_counter()
print("Wall time:", t1 - t0)
```
